```python
#12. Find the trips that are longer than previous trip. Tip: Luckily, trips are sorted by date and trip IDs are consistent 
#with date. 


q = """
         SELECT t1.trip_id as previous_trip, t2.trip_id as next_trip
         FROM df t1
         join 
         (SELECT trip_id from df) as t2
         on t1.trip_id = t2.trip_id-1
         
     """

# print(pysqldf(q).dtypes)
pysqldf(q).head(10)
  
```

```python
## Now add trip duration to the trip ids to calculate trip duration in seconds

q = """
         SELECT t1.trip_id as previous_trip, t2.trip_id as next_trip,
         CAST(strftime('%s', t1.Lpep_dropoff_datetime) as second) - CAST(strftime('%s', t1.lpep_pickup_datetime) as second) as previous_trip_duration,
         t2.next_trip_duration
         FROM df t1
         join 
         (SELECT trip_id, Lpep_dropoff_datetime, lpep_pickup_datetime,
         CAST(strftime('%s', Lpep_dropoff_datetime) as second) - CAST(strftime('%s', lpep_pickup_datetime) as second) as next_trip_duration
         from df) as t2
         on t1.trip_id = t2.trip_id-1
         
     """

print(pysqldf(q).dtypes)
pysqldf(q).head(10)
```

```python
## Second way to above query
q = """
         SELECT t1.trip_id as previous_trip, t2.trip_id next_trip,
         CAST(strftime('%s', t1.Lpep_dropoff_datetime) as second) - CAST(strftime('%s', t1.lpep_pickup_datetime) as second) as previous_trip_duration,
         CAST(strftime('%s', t2.Lpep_dropoff_datetime) as second) - CAST(strftime('%s', t2.lpep_pickup_datetime) as second) as next_trip_duration
         FROM df t1
         join 
         (select trip_id, Lpep_dropoff_datetime, lpep_pickup_datetime from df) as t2
         on t1.trip_id = t2.trip_id-1
         
     """

# print(pysqldf(q).dtypes)
pysqldf(q).head(10)
```

```python
## It is time to compare the trip durations

q = """
         SELECT t1.trip_id as previous_trip, t2.trip_id next_trip,
         CAST(strftime('%s', t1.Lpep_dropoff_datetime) as second) - CAST(strftime('%s', t1.lpep_pickup_datetime) as second) as previous_trip_duration,
         CAST(strftime('%s', t2.Lpep_dropoff_datetime) as second) - CAST(strftime('%s', t2.lpep_pickup_datetime) as second) as next_trip_duration
         FROM df t1
         join 
         (select trip_id, Lpep_dropoff_datetime, lpep_pickup_datetime from df) as t2
         on t1.trip_id = t2.trip_id-1
         where
         previous_trip_duration < next_trip_duration
         
     """

# print(pysqldf(q).dtypes)
pysqldf(q).head(10)

## Now we see only trip that are longer than the previous one.

```

```python
## Step 3: adding rank criteria into "on" condition. Note that we subtract 1 so that predecessor and successor trips would be
## side by side. I modifed the names of columns so that the comparison would be easier.

# No need for all columns here. Keeping them to make the table more understandable.

q = """
        WITH duration_ranking as(
        SELECT driver_id, lpep_pickup_datetime,
                CAST(strftime('%s', Lpep_dropoff_datetime) as second) - CAST(strftime('%s', lpep_pickup_datetime) as second) as trip_duration,
                RANK() OVER(PARTITION BY driver_id ORDER BY lpep_pickup_datetime) rank
        FROM df)
        
        SELECT t1.driver_id, t1.lpep_pickup_datetime as time_previous, t2.lpep_pickup_datetime as time_next, 
               t1.trip_duration as duration_previous, t2.trip_duration as duration_next, t1.rank as rank_previous, t2.rank as rank_next 
        FROM duration_ranking t1
        JOIN duration_ranking t2
        ON t1.driver_id = t2.driver_id 
        AND t1.rank = t2.rank-1
        
    """



pysqldf(q).head(20)
```

```python
# Look at the column difference by creating subquery

q = """
        WITH duration_ranking as(
        SELECT driver_id, lpep_pickup_datetime,
                CAST(strftime('%s', Lpep_dropoff_datetime) as second) - CAST(strftime('%s', lpep_pickup_datetime) as second) as trip_duration,
                RANK() OVER(PARTITION BY driver_id ORDER BY lpep_pickup_datetime) rank
        FROM df)
        
        SELECT *, duration_next - duration_previous as duration_difference
        FROM(
        
        SELECT t1.driver_id, t1.lpep_pickup_datetime as time_previous, t2.lpep_pickup_datetime as time_next, 
               t1.trip_duration as duration_previous, t2.trip_duration as duration_next, t1.rank as rank_previous, t2.rank as rank_next 
        FROM duration_ranking t1
        JOIN duration_ranking t2
        ON t1.driver_id = t2.driver_id 
        AND t1.rank = t2.rank-1)
        
    """



pysqldf(q).head(20)
```
